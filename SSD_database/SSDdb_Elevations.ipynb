## 1. Read in Elevations data

```python
def read_elevations(csvfile, acct = 'Checking'):
    dfename = pd.read_csv(pathElevations + csvfile, skiprows=3)
    dfename['SourceFile'] = csvfile
    dfename['Account'] = acct
    global dfe
    dfe = dfe.append(dfename, ignore_index=True)

dfe = pd.DataFrame()
read_elevations('Elevations_2015.csv')
read_elevations('Elevations_2014.csv')
read_elevations('Elevations_2013.csv')
read_elevations('Elevations_2012.csv')
read_elevations('Elevations_2011.csv')
read_elevations('Elevations_Savings.csv', 'Savings')

dfe1 = dfe.copy()
dfe.shape # (489, 11)
```

## 2. Preprocessing

```python
dfe = dfe1.copy()

dfe.columns = map(str.strip, dfe.columns)
dfe.rename(columns={
    'Transaction Number' : 'TransID',
    'Amount Debit' : 'Debit',
    'Amount Credit' : 'Credit',
    'Description' : 'El_Description',
    'Memo' : 'El_Memo',
    'Check Number' : 'El_CheckNbr'
    }, inplace=True)
cols = ['Date', 'SourceFile', 'Account', 'TransID',
    'Debit', 'Credit', 'Balance', 'El_Description', 'El_Memo', 'El_CheckNbr']
dfe = dfe[cols]
dfe[['El_Description']] = dfe[['El_Description']].astype(str)
dfe[['El_Memo']] = dfe[['El_Memo']].astype(str)
dfe['El_CheckNbr'] = dfe['El_CheckNbr'].fillna(0.0).astype(int).astype(str)
dfe[['Debit']] = dfe[['Debit']].astype(float)
dfe['Debit'].fillna(0, inplace=True)
dfe[['Credit']] = dfe[['Credit']].astype(float)
dfe['Credit'].fillna(0, inplace=True)
dfe['Amount'] = dfe['Debit'] + dfe['Credit']
dfe['Date'] = pd.to_datetime(dfe['Date'], format='%m/%d/%Y')
dfe['For Date'] = dfe['Date']
dfe['For Month'] = dfe['For Date'].apply(lambda dt: int(dt.strftime('%Y%m')))
dfe['Entries'] = 1
#dfe_pre2013 = dfe[dfe['Date'] < '01-01-2013']
#dfe = dfe[dfe['Date'] >= '01-01-2013']

dfe2 = dfe.copy()
dfe.shape # (489, 14)
```

## 3. Assign who, AcctNm categories

```python
dfe = dfe2.copy()

def assign_cats_el(s):
    '''Assign descriptors for every transaction:
    who (Paypal, Square, member, vendor, etc.)
    what1: what2: AcctNm: . . . '''
    
    who = 'UNKNOWN'
    AcctNm = 'UNKNOWN'
    
    ########## Dues, Donations, Dividends, some Fees ##########
    if ('SQUARE' in s['El_Memo'].upper()) | \
        ('SQC' in s['El_Memo'].upper()) | \
        ('SQUARE' in s['El_Description'].upper()):
        who = 'Square'
        AcctNm = 'Deposits'
    elif ('PAYPAL' in s['El_Memo'].upper()) | \
        ('PAYPAL' in s['El_Description'].upper()):
        who = 'Paypal'
        AcctNm = 'Transfers'
    elif ('PP' in s['El_Memo'].upper()) & \
        (s['El_Description'] == 'Withdrawal'):
        who = 'Paypal'
        AcctNm = 'Fees Paypal Other'
    elif 'DEPOSIT ADJUSTMENT' in s['El_Description'].upper():
        who = 'Geoffrey Terrell'
        AcctNm = 'Dues Monthly'
    elif 'HOME BANKING' in s['El_Description'].upper():
        who = 'Self'
        AcctNm = 'Transfers'
    elif 'DIVIDEND' in s['El_Description'].upper():
        who = 'Elevations'
        AcctNm = 'Dividends'
    elif 'DEPOSIT' in s['El_Description'].upper():
        AcctNm = 'Deposits'
    
    ########## Fees ##########
    elif s['El_Description'] == 'Business Fee':
        who = 'Elevations'
        AcctNm = 'Fees Bank Monthly'
    elif (s['El_Description'] == 'Courtesy Pay Fee') | \
        ('ITEM FEE STALE DATE' in s['El_Memo'].upper()):
        # Courtesy pay fee for? not sure why
        who = 'Elevations'
        AcctNm = 'Fees Bank Other'
    
    ########## Rent and Utilities ##########
    elif (s['El_Description'] == 'Withdrawal by Check') & \
        (s['For Month'] >= 201202) & (s['For Month'] <= 201204):
        who = 'Westland'
        AcctNm = 'Rent'
    elif ('Draft' in s['El_Description']) & (s['For Month'] <= 201303) & \
        (s['Amount'] == -1250):
        who = 'Westland'
        AcctNm = 'Rent'
    elif s['El_CheckNbr'] in ['127', '157', '158', '159', '177', '179']:
        who = 'Westland'
        AcctNm = 'Utilities'
    elif s['El_CheckNbr'] in ['151', '180', '181', '182', '183']:
        who = 'Westland'
        AcctNm = 'Rent and Utilities'
    
    elif (s['El_CheckNbr'] == '184') | \
        ((s['Date'] == '04/03/2014') & (s['Amount'] == -61.46)):
        who = 'Kreizel'
        AcctNm = 'Rent'
    elif (s['El_CheckNbr'] == '227') | \
        (('KREIZEL' in s['El_Description'].upper()) & (s['For Month'] == 201406)):
        who = 'Kreizel'
        AcctNm = 'Utilities'
    elif (s['El_CheckNbr'] in ['226', '228', '229', '231']) | \
        ('KREIZEL' in s['El_Description'].upper()) | \
        ('KREIZEL' in s['El_Memo'].upper()):
        who = 'Kreizel'
        AcctNm = 'Rent and Utilities'
    
    ########## Internet, Trash ##########
    elif ('LIVE WIRE' in s['El_Memo'].upper()) | \
        ('LIVE WIRE' in s['El_Description'].upper()) | \
        (s['El_CheckNbr'] == '101'):
        who = 'Live Wire'
        AcctNm = 'Internet'
    elif 'WESTERN DISPOSAL' in s['El_Description'].upper():
        who = 'Western Disposal'
        AcctNm = 'Trash'
    
    ########## Special Items ##########
    elif s['El_CheckNbr'] == '152':
        who = 'Mill'
        AcctNm = 'Mill'
    
    ########## Misc Expenses ##########
    elif 'Withdrawal' in s['El_Description']:
        if 'ITEM STALE DATE' in s['El_Memo'].upper():
            who = 'Self'
            AcctNm = 'Fees Other'
        elif 'SOS REGISTRATION' in s['El_Memo'].upper():
            who = 'CO Sec of State'
            AcctNm = 'SOS Registration'
                
    return pd.Series({
        'who' : who,
        'AcctNm' : AcctNm})


# dfe = dfe2.copy()
dfe_cats = dfe.apply(assign_cats_el, axis=1)
dfe = dfe.join(dfe_cats)

dfe3 = dfe.copy()
dfe.shape # (489, 16)
```

## 4. Split AcctNm = 'Rent and Utilities', reassign For Month

```python
dfe = dfe3.copy()

def split_rentutil(s):
    if s['For Month'] < 201304:
        rent = -1250
        utilities = s['Debit'] - rent
    elif s['For Month'] < 201308:
        utilities = -150
        rent = s['Debit'] - utilities
    elif s['Debit'] > -200:
        utilities = 0
        rent = s['Debit']
    else:
        utilities = -200
        rent = s['Debit'] - utilities
    return pd.Series({'Rent': rent, 'Utilities': utilities})

dfe_not_ru = dfe[dfe['AcctNm'] != 'Rent and Utilities']

dfe_rent = dfe.copy()
dfe_rent = dfe_rent[dfe_rent['AcctNm'] == 'Rent and Utilities']
dfe_rent['AcctNm'] = 'Rent'
dfe_rent['Debit'] = dfe_rent.apply(split_rentutil, axis=1)['Rent']

dfe_util = dfe.copy()
dfe_util = dfe_util[dfe_util['AcctNm'] == 'Rent and Utilities']
dfe_util['AcctNm'] = 'Utilities'
dfe_util['Debit'] = dfe_util.apply(split_rentutil, axis=1)['Utilities']

dfe = pd.concat([dfe_not_ru, dfe_rent, dfe_util])
dfe['Amount'] = dfe['Debit'] + dfe['Credit']

#dfe_rent[['Date','AcctNm','who','Amount']].sort('Date')
#dfe_util[['Date','AcctNm','who','Amount']].sort('Date')

# reassign For Date for rent and utilities
dfe_not_ru = dfe[(dfe['AcctNm'] != 'Rent') & (dfe['AcctNm'] != 'Utilities')]
dfe_ru = dfe.copy()
dfe_ru = dfe_ru[(dfe_ru['AcctNm'] == 'Rent') | (dfe_ru['AcctNm'] == 'Utilities')]
dfe_ru['For Date'] = [d + pd.DateOffset(months=1) if int(d.strftime('%d')) > 24 else d for d in dfe_ru['Date']]
dfe_ru['For Date'] = [d + pd.DateOffset(months=1) if int(d.strftime('%Y%m%d')) == 20130904 else d for d in dfe_ru['For Date']]
dfe_ru['For Date'] = [d - pd.DateOffset(months=1) if int(d.strftime('%Y%m%d')) == 20140529 else d for d in dfe_ru['For Date']]
dfe_ru['For Month'] = dfe_ru['For Date'].apply(lambda dt: int(dt.strftime('%Y%m')))

dfe = pd.concat([dfe_not_ru, dfe_ru])

dfe4 = dfe.copy()
dfe.shape # (516, 16)
```

## 5. Split AcctNm = 'Deposits' by using RevenueDetail.csv
- first need to reconcile totals by Date, then for those that match, substitute detail

```python
dfe = dfe4.copy()

# preprocessing: add fields needed for Paypal merging
dfe['Attendees'] = 0
dfe['Dues_Disc'] = 0
dfe['Dues_Rate'] = 0
dfe['Mbrs'] = 0
dfe['Mbrs_Reg'] = 0
dfe['Mbrs_SS'] = 0
dfe['Mbrs_Fam'] = 0
dfe['Mbrs_UNK'] = 0

dfe_dd = dfe[dfe['AcctNm'] == 'Deposits']
dfe_nodd = dfe[dfe['AcctNm'] != 'Deposits']

dfe_dd_bydt = dfe_dd.copy()
dfe_dd_bydt = dfe_dd_bydt[['Date', 'Amount', 'Debit', 'Credit']]
dfe_dd_bydt = dfe_dd_bydt.groupby('Date').sum()

# Use Revenue Detail from spreadsheet: dfrd_el, dfrd_el_bydt

# summarize by month, merge to Elevations data and reconcile
dd_compare = pd.merge(
    dfrd_el_bydt.reset_index(), 
    dfe_dd_bydt.reset_index(), 
    how='outer', on='Date', sort = 'TRUE')

dd_compare.drop('index', axis=1, inplace=True)
dd_compare.columns = ['Date', 'Spreadsheet', 'Elevations', 'ElevationsDr', 'ElevationsCr']
dd_compare.fillna(0, inplace=True)
dd_compare['Diff'] = dd_compare['Spreadsheet'] - dd_compare['Elevations']

# check for differences, do not have detail before 2013, and 2 dates in 2013 (2/1, 3/22)
dd_compare[(dd_compare['Diff'] != 0) & (dd_compare['Elevations'] != 0) & (dd_compare['Date'] > '12-31-2012')]
```

```python
# Next, for dates that matched and 0 diff, substitute detail rev data
dd_subs_datelist = list(dd_compare[(dd_compare['Diff'] == 0) & \
    (dd_compare['Elevations'] != 0) & \
    (dd_compare['Date'] > '12-31-2012')]['Date'])

dfe_nosubs = dfe_dd[~dfe_dd['Date'].isin(dd_subs_datelist)]
dfe_subs1 = dfe_dd[dfe_dd['Date'].isin(dd_subs_datelist)]

# Reduce so there is only one row per Date and Account, before merging to detail by date
dfekeep = ['Date', 'Account', 'Entries']
dfe_subs2 = dfe_subs1[dfekeep].drop_duplicates()
dfrdkeep = ['Date', 'AcctNm', 'Amount', 'who', 'how', 'For Date', 'Qty']

dfe_subs3 = pd.merge(dfe_subs2, dfrd_el[dfrdkeep], on='Date', suffixes = ('', '_y')) 
#dfe_subs3.rename(columns={'Category' : 'AcctNm'}, inplace=True)
dfe_subs3['Qty'].fillna(1, inplace=True)
dfe_subs3['Debit'] = [d if d < 0 else 0 for d in dfe_subs3['Amount']]
dfe_subs3['Credit'] = [d if d >= 0 else 0 for d in dfe_subs3['Amount']]

# Now, derive For Month and assign membership columns
# Inputs: Date, Account, Entries, AcctNm, who, how, For Date, Qty, Debit, Credit, Amount
def assign_dddtl(s):
    For_Month = 0
    Attendees = 0
    Dues_Rate = 0
    Mbrs = 0
    Mbrs_Reg = 0
    Mbrs_SS = 0
    Mbrs_Fam = 0
    Mbrs_UNK = 0
    Dues_Disc = 0
    
    if pd.isnull(s['For Date']): 
        For_Month = int(s['Date'].strftime('%Y%m'))
    else: 
        For_Month = int(s['For Date'].strftime('%Y%m'))
    
    if s['AcctNm'] == 'Workshops':
        Attendees = max(1,s['Qty'])
    elif s['AcctNm'] == 'Dues Monthly':
        Mbrs = 1
        if (For_Month == 201412) & (s['who'] == 'John West'):
            # 12/1/14 dues split over 2 months
            Mbrs = 0.5
            Mbrs_SS = 0.5
            Dues_Rate = 25.0
        elif ((s['Amount'] >= 11) & (s['Amount'] <= 13)) | \
            ((s['Amount'] >= 23) & (s['Amount'] <= 25)) | \
            (s['Amount'] == 30) | (s['Amount'] == 40):
            Mbrs_SS = 1
            Dues_Rate = 25.0
        elif ((s['Amount'] >= 35) & (s['Amount'] <= 38)) | \
            ((s['Amount'] >= 62) & (s['Amount'] <= 65)) | \
            ((s['Amount'] >= 71) & (s['Amount'] <= 75)):
            Mbrs_Reg = 1
            if (s['Amount'] >= 62) & (s['Amount'] <= 65):
                Dues_Rate = 65.0
            else:
                Dues_Rate = 75.0
        elif ((s['Amount'] >= 48) & (s['Amount'] <= 50)) | \
            ((s['Amount'] >= 95) & (s['Amount'] <= 100)):
            Mbrs_Fam = 1
            Dues_Rate = 100.0
        else:
            Mbrs_UNK = 1
        
        if (Mbrs == 1) & \
            (((s['Amount'] >= 11) & (s['Amount'] <= 13)) | \
            ((s['Amount'] >= 31) & (s['Amount'] <= 33)) | \
            ((s['Amount'] >= 35) & (s['Amount'] <= 38)) | \
            ((s['Amount'] >= 46) & (s['Amount'] <= 50))):
            Dues_Disc = 1
        
        if (s['who'] == 'Ryan Bennett') & ((For_Month == 201308) | (For_Month == 201309)):
            Mbrs_Reg = 1.0
            Mbrs_SS = 0
            Dues_Rate = 75.0
            Dues_Disc = 1
    
    return pd.Series({
        'For Month' : For_Month,
        'Attendees' : Attendees, 
        'Dues_Rate': Dues_Rate,
        'Mbrs' : Mbrs,
        'Mbrs_Reg' : Mbrs_Reg,
        'Mbrs_SS' : Mbrs_SS,
        'Mbrs_Fam' : Mbrs_Fam,
        'Mbrs_UNK' : Mbrs_UNK,
        'Dues_Disc' : Dues_Disc})


dfe_subs3b = dfe_subs3.apply(assign_dddtl, axis=1)
dfe_subs4 = dfe_subs3.join(dfe_subs3b)

dfe = pd.concat([dfe_nodd, dfe_nosubs, dfe_subs4])

dfe5 = dfe.copy()
dfe.shape # (666, 26)
```

```python
dfe = dfe5.copy()

dfe_exp = dfe[dfe['AcctNm'] == 'UNKNOWN']
dfe_noexp = dfe[dfe['AcctNm'] != 'UNKNOWN'] 

dfe_exp_bydt = dfe_exp.copy()
dfe_exp_bydt = dfe_exp_bydt[['Date', 'Amount', 'Debit', 'Credit']]
dfe_exp_bydt = dfe_exp_bydt.groupby('Date').sum()

#dfe_exp_bydt = dfe_exp[['Date', 'Amount', 'Debit', 'Credit']].groupby('Date').sum()

# Use Expense Detail from spreadsheet: dfed_el, dfed_el_bydt

# summarize by month, merge to Elevations data and reconcile
exp_compare = pd.merge(
    dfed_el_bydt.reset_index(), 
    dfe_exp_bydt.reset_index(), 
    how='outer', on='Date', sort = 'TRUE')

exp_compare.drop('index', axis=1, inplace=True)
exp_compare.columns = ['Date', 'Spreadsheet', 'Elevations', 'ElevationsDr', 'ElevationsCr']
exp_compare.fillna(0, inplace=True)
exp_compare['Diff'] = exp_compare['Spreadsheet'] - exp_compare['Elevations']

# check for differences, none nonzero in Elevations with different amount in spreadsheet
exp_compare[(exp_compare['Diff'] != 0) & (exp_compare['Elevations'] != 0)]# & (exp_compare['Date'] > '12-31-2011')]
```

```python
# Next, for dates that matched and 0 diff, substitute detail rev data
exp_subs_datelist = list(exp_compare[(exp_compare['Diff'] == 0) & \
    (exp_compare['Elevations'] != 0) & \
    (exp_compare['Date'] > '12-31-2012')]['Date'])

dfe_nosubs = dfe_exp[~dfe_exp['Date'].isin(exp_subs_datelist)]
dfe_subs1 = dfe_exp[dfe_exp['Date'].isin(exp_subs_datelist)]

# Reduce so there is only one row per Date and Account, before merging to detail by Date
dfekeep = ['Date', 'Account', 'Entries']
dfe_subs2 = dfe_subs1[dfekeep].drop_duplicates()
dfedkeep = ['Date', 'AcctNm', 'Amount', 'who', 'how', 'For Date', 'CheckNbr']

dfe_subs3 = pd.merge(dfe_subs2, dfed_el[dfedkeep], on='Date', suffixes = ('', '_y')) 
dfe_subs3['Debit'] = [d if d < 0 else 0 for d in dfe_subs3['Amount']]
dfe_subs3['Credit'] = [d if d >= 0 else 0 for d in dfe_subs3['Amount']]

# Get For Month
def assign_dddtl(s):
    ''' Assign dues / donations detail fields. Note set it up
    so that payments for multiple months are in separate rows.'''
    For_Month = 0
    
    if pd.isnull(s['For Date']): 
        For_Month = int(s['Date'].strftime('%Y%m'))
    else: 
        For_Month = int(s['For Date'].strftime('%Y%m'))
    
    return pd.Series({'For Month' : For_Month})


dfe_subs3b = dfe_subs3.apply(assign_dddtl, axis=1)
dfe_subs4 = dfe_subs3.join(dfe_subs3b)

dfe = pd.concat([dfe_noexp, dfe_nosubs, dfe_subs4])

# cleanup: fill in nas, keep only necessary well-populated columns
dfe['For Date'] = dfe['For Month'].apply(lambda x: datetime.date(int(x/100), int(x - int(x/100)*100), 1))
nacols = ['Attendees', 'Dues_Disc', 'Dues_Rate', 'Mbrs', 'Mbrs_Reg', 'Mbrs_SS', 'Mbrs_Fam', 'Mbrs_UNK']
for col in nacols:
    dfe[col] = dfe[col].fillna(0)
keepcols = ['For Month', 'For Date', 'Date', 'who', 'AcctNm', 'Account', 
            'Debit', 'Credit', 'Amount', 'Entries', 'Attendees', 
            'Dues_Disc', 'Dues_Rate', 'Mbrs', 'Mbrs_Reg', 'Mbrs_SS', 'Mbrs_Fam', 'Mbrs_UNK']
dfe = dfe[keepcols]

dfe6 = dfe.copy()
dfe.shape # (666, 18)
```
